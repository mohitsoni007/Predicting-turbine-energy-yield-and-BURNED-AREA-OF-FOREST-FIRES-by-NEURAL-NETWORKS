```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
dataset['size_category']= label_encoder.fit_transform(dataset['size_category']) 
dataset
```

```python
def norm_data(i):
    x =((i- i.min())/(i.max()-i.min()))
    return x
normalized_data =norm_data(dataset.iloc[:,2:-1])
```

```python
normalized_data
```

```python
X=normalized_data
y=dataset['size_category']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size =0.3,random_state =42)
```

```python
features =X.shape[1]
features
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
optmizer =RMSprop(0.001)#0.01 is a learning rate
model =keras.Sequential([
    keras.layers.Dense(12,input_dim =28,activation='relu'),
    keras.layers.Dense(50,activation ='sigmoid'),
    keras.layers.Dense(1,activation='sigmoid')
])
model.compile(optimizer =optmizer,loss= 'binary_crossentropy',metrics =['accuracy'])
```

```python
seed_val =42
import random
random.seed(seed_val)
tensorflow.random.set_seed(seed_val)
```

```python
model.fit(X_train,y_train,epochs =10)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred = model.predict(X_test)
```
